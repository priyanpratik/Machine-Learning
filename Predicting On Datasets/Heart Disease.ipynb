```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("heart_disease.csv")
df.head(20)
```

```python
print(df.isnull().sum())
```

```python
# Data Cleaning
# Calcuating mean or median of columns having NaN values

education=df.education.median()
cigsPerDay=df.cigsPerDay.median()
totChol=df.totChol.mean()
glucose=df.glucose.mean()
BMI=df.BMI.mean()
BPMeds=df.BPMeds.median()
print(education,cigsPerDay,totChol,glucose,BMI,BPMeds)
```

```python
# Filling NaN values with mean or median value

df.education.fillna(education,inplace=True)
df.cigsPerDay.fillna(cigsPerDay,inplace=True)
df.totChol.fillna(totChol,inplace=True)
df.glucose.fillna(glucose,inplace=True)
df.BMI.fillna(BMI,inplace=True)
df.BPMeds.fillna(BPMeds,inplace=True)
print(df.isnull().sum())
```

```python
# dropping 1 row having NaN in heartRate column
df.dropna(inplace=True)
```

```python
print(df.isnull().sum())
```

```python
df
```

```python
y=np.array(df["TenYearCHD"])
del df["TenYearCHD"]
print(y)
```

```python
x=df.values
print(x)
```

```python
# splitting data for training the model and testing

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y)
print(x_train.shape,y_train.shape,x_test.shape)
```

```python
# using Logistic Regression algorithm for predicting
from sklearn.linear_model import LogisticRegression
algo=LogisticRegression(max_iter=1000,solver="liblinear",C=2)
algo.fit(x_train,y_train)
y_pred=algo.predict(x_test)

# calculating score
pred_score=algo.score(x_test,y_test)
train_score=algo.score(x_train,y_train)
print(pred_score,train_score)
```
