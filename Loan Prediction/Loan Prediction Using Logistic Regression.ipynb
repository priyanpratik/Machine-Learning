```python
import pandas as pd
import numpy as np
```

```python
loan=pd.read_csv("Loan_Prediction.csv")
loan
```

```python
del loan["Loan_ID"]
del loan["CoapplicantIncome"]
```

```python
loan
```

```python
def fun(s):
    if(s=="Male"):
        return 1
    elif(s=="Female"):
        return 0

loan["Gender"]=loan.Gender.apply(fun)
loan
```

```python
def fun(s):
    if(s=="Yes"):
        return 1
    elif(s=="No"):
        return 0

loan["Married"]=loan.Married.apply(fun)
loan
```

```python
def fun(s):
    if(s=="Graduate"):
        return 1
    elif(s=="Not Graduate"):
        return 0

loan["Education"]=loan.Education.apply(fun)
loan
```

```python
def fun(s):
    if(s=="Yes"):
        return 1
    elif(s=="No"):
        return 0

loan["Self_Employed"]=loan.Self_Employed.apply(fun)
loan
```

```python
def fun(s):
    if(s=="Urban"):
        return 2
    elif(s=="Rural"):
        return 1
    elif(s=="Semiurban"):
        return 0

loan["Property_Area"]=loan.Property_Area.apply(fun)
loan
```

```python
def fun(s):
    if(s=="Y"):
        return 1
    elif(s=="N"):
        return 0

loan["Loan_Status"]=loan.Loan_Status.apply(fun)
loan
```

```python
def fun(s):
    if(s=="3+"):
        return 3
    elif(s=="1"):
        return 1
    elif(s=="0"):
        return 0
    elif(s=="2"):
        return 2

loan["Dependents"]=loan.Dependents.apply(fun)
loan
```

```python
loan.isna().sum()
```

```python
# calculating mean and median for NAN values
```

```python
gender=loan["Gender"].median()
married=loan["Married"].median()
dependents=loan["Dependents"].median()
selfemp=loan["Self_Employed"].median()
loanamount=loan["LoanAmount"].mean()
term=loan["Loan_Amount_Term"].mean()
credhis=loan["Credit_History"].median()
gender,married,dependents,selfemp,loanamount,term,credhis
```

```python
loan.shape
```

```python
import math
for i in range(614):
    x=float(loan.Gender[i])
    if(math.isnan(x)):
        loan.Gender[i]=gender


for i in range(614):
    x=float(loan.Married[i])
    if(math.isnan(x)):
        loan.Married[i]=married
        

for i in range(614):
    x=float(loan.Dependents[i])
    if(math.isnan(x)):
        loan.Dependents[i]=dependents
        

for i in range(614):
    x=float(loan.Self_Employed[i])
    if(math.isnan(x)):
        loan.Self_Employed[i]=selfemp
        
        
for i in range(614):
    x=float(loan.LoanAmount[i])
    if(math.isnan(x)):
        loan.LoanAmount[i]=loanamount
        
        
for i in range(614):
    x=float(loan.Loan_Amount_Term[i])
    if(math.isnan(x)):
        loan.Loan_Amount_Term[i]=term
        

for i in range(614):
    x=float(loan.Credit_History[i])
    if(math.isnan(x)):
        loan.Credit_History[i]=credhis
```

```python
loan.isna().sum()
```

```python
Y=np.array(loan["Loan_Status"])
del loan["Loan_Status"]
Y
```

```python
X=loan.values
X.shape
```

```python
# splitting into testing and training
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,random_state=0)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression()
clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)
y_pred
```

```python
print(clf.score(x_test,y_test))
```
