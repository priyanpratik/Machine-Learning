```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("pima-data.csv")
data
```

```python
# Checking NAN Values
```

```python
data.isna().sum()
```

```python
def fun(s):
    if(s==True):
        return 1
    elif(s==False):
        return 0
    
data["diabetes"]=data.diabetes.apply(fun)
data
```

```python
# storing output in seperate array (Y) ie. diabetes column is stored in Y
Y=np.array(data["diabetes"])

# Now we will delete the diabetes column from dataset
del data["diabetes"]
Y.shape
```

```python
# storing rest 9 columns in X which is basically input
X=data.values
X.shape
```

```python
# model_selection is used to distribute data into training and testing
from sklearn.model_selection import train_test_split


x_train,x_test,y_train,y_test=train_test_split(X,Y,random_state=0)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression


# clf is an object of LogisticRegression

# max_iter runs loop to converge to a single point
clf=LogisticRegression(max_iter=500)

clf.fit(x_train,y_train)
y_predict=clf.predict(x_test)
```

```python
from sklearn import metrics

metrics.accuracy_score(y_test, y_predict)
```
